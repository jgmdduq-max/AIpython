```python
# 가위바위보 #산지니 수정본
import random

hand1 = random.randint(1, 3)
hand2 = random.randint(1, 3)

# 1:가위, 2:바위, 3:보
if hand1 == 1:
    print('사람1:가위')
elif hand1 == 2:
    print('사람1:바위')
else:
    print('사람1:보')

if hand2 == 1:
    print('사람2:가위')
elif hand2 == 2:
    print('사람2:바위')
else:
    print('사람2:보')  # (기존 코드: 사람3 오타)

# 승패 판단
if hand1 == hand2:
    print('비겼습니다.')
elif (hand1 == 1 and hand2 == 3) or (hand1 == 2 and hand2 == 1) or (hand1 == 3 and hand2 == 2):
    print('사람 1이 이겼습니다.')
else:
    print('사람 2가 이겼습니다.')
```
