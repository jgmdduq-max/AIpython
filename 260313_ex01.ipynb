```python
 a = 254638214           # 첫 번째 숫자를 저장할 변수(상자)를 만든다. (예: 10)
b = 3545454654             # 두 번째 숫자를 저장할 변수를 만든다. (예: 3)

add = a + b       # 덧셈(+) 결과를 저장한다. a와 b를 더한 값을 add 상자에 넣는다.
sub = a - b       # 뺄셈(-) 결과를 저장한다. a에서 b를 뺀 값을 sub에 넣는다.
mul = a * b       # 곱셈(*) 결과를 저장한다. a와 b를 곱한 값을 mul에 넣는다.
div = a / b       # 나눗셈(/) 결과를 저장한다. a를 b로 나눈 값을 div에 넣는다(소수점 나올 수 있음).

print(add)        # 덧셈 결과를 화면에 출력한다.
print(sub)        # 뺄셈 결과를 화면에 출력한다.
print(mul)        # 곱셈 결과를 화면에 출력한다.
print(div)        # 나눗셈 결과를 화면에 출력한다.
```

```python
a = input("첫 번째 숫자를 입력하세요: ")   # 키보드로 입력을 받는다(하지만 지금은 '글자' 상태로 들어온다).
b = input("두 번째 숫자를 입력하세요: ")   # 두 번째 입력도 받는다(이것도 '글자' 상태).

a = int(a)                               # 사칙연산을 하려면 숫자가 필요하니까, a를 정수(int)로 바꾼다.
b = int(b)                               # b도 정수(int)로 바꾼다.

add = a + b                              # 덧셈(+) 결과를 계산해서 add에 저장한다.
sub = a - b                              # 뺄셈(-) 결과를 계산해서 sub에 저장한다.
mul = a * b                              # 곱셈(*) 결과를 계산해서 mul에 저장한다.
div = a / b                              # 나눗셈(/) 결과를 계산해서 div에 저장한다(소수 가능).

print(add)                               # 덧셈 결과를 출력한다.
print(sub)                               # 뺄셈 결과를 출력한다.
print(mul)                               # 곱셈 결과를 출력한다.
print(div)                               # 나눗셈 결과를 출력한다.
```
