```python
# 입력: 분(minute)을 몇 분 입력받을지
n = int(input("분을 몇 번 입력받을까요? : "))

for _ in range(n):
    m = int(input("분을 입력하세요: "))

    hours = m // 60      # 60으로 나눈 몫 = 시간
    minutes = m % 60     # 60으로 나눈 나머지 = 분

    print(f"{m}분은 {hours}시간 {minutes}분입니다.")
```

```python
n = int(input("분을 몇 번 입력받을까요? : "))

for _ in range(n):
    m = int(input("분을 입력하세요: "))

    hours = m // 60      # 60으로 나눈 몫 = 시간
    minutes = m % 60     # 60으로 나눈 나머지 = 분

    print(f"{m}분은 {hours}시간 {minutes}분입니다.")
```

```python
# 사용자에게 숫자를 입력받습니다.
num = int(input("숫자를 입력하세요: "))

# 2로 나눈 나머지가 0이면 짝수, 아니면 홀수입니다.
if num > 0:
    print(f"{num}는(은) 양수입니다.")
else:
    print(f"{num}는(은) 음수입니다.")
```
